```python
import numpy as np
import cupy as cp
import astropy.units as u
from astropy.io import fits
from matplotlib.patches import Rectangle, Circle
from pathlib import Path
from IPython.display import clear_output
from importlib import reload

from roman_cgi_iefc_2.poppy_roman_cgi_phasec import cgi
from roman_cgi_iefc_2 import iefc_utils as iefcu
from roman_cgi_iefc_2 import pwp, efc

import misc

data_dir = Path('/groups/douglase/kians-data-files/roman-cgi-iefc-data')
flatmaps_dir = Path.home()/'src/pyfalco/roman/flatmaps'

dm1_flatmap = fits.getdata(flatmaps_dir/'dm1_m_flat_hlc_band1.fits')
dm2_flatmap = fits.getdata(flatmaps_dir/'dm2_m_flat_hlc_band1.fits')
dm1_design = fits.getdata(flatmaps_dir/'dm1_m_design_hlc_band1.fits')
dm2_design = fits.getdata(flatmaps_dir/'dm2_m_design_hlc_band1.fits')

hlci = cgi.CGI_POPPY(polaxis=0, use_opds=True, use_fieldstop=True, quiet=False, ngpus=1/4)
misc.myimshow2(hlci.inwave.intensity, hlci.inwave.phase)

hlci.set_dm1(2*dm1_flatmap)
hlci.set_dm2(2*dm2_flatmap)

ref_psf = hlci.calc_psf()[-1]
misc.myimshow(ref_psf.intensity, lognorm=True)

npsf = hlci.npsf
Nact = hlci.Nact
```

```python
hlci = cgi.CGI_POPPY(polaxis=0, use_opds=False, use_fieldstop=False, quiet=False, ngpus=1/4)
ref_psf = hlci.calc_psfs()[0][-1]
misc.myimshow(ref_psf.intensity, lognorm=True)
```

```python
xfp = np.linspace(-0.5, 0.5, npsf) * npsf * hlci.psf_pixelscale_lamD
xf,yf = np.meshgrid(xfp,xfp)

edge = 1
iwa = 3
owa = 9

# Create the mask that is used to select which region to make dark.
dh_params = {
    'inner_radius' : iwa,
    'outer_radius' : owa,
    'edge_position' : edge,
    'direction' : '+x'
}
dh_mask = iefcu.create_annular_focal_plane_mask(xf, yf, dh_params).ravel()

mask = cp.array(dh_mask).reshape(npsf,npsf)
nmask = dh_mask.sum()

misc.myimshow(dh_mask.reshape(npsf,npsf), pxscl=hlci.psf_pixelscale_lamD)
```

```python
dm = cp.ones((Nact**2))*1e-9

E = jac.dot(dm)
E = E[:nmask] + 1j*E[nmask:]

I = cp.zeros((npsf,npsf))
cp.place(I, mask=dh_mask, vals=abs(E)**2)

misc.myimshow2(I*190*2, ref_psf.intensity*mask, lognorm1=True, lognorm2=True)
misc.myimshow(abs(I*190*2/ref_psf.intensity*mask), lognorm=True)
```

```python
probe_amp = 0.01*hlci.wavelength_c.value

probe_radius = 15
probe_offset = (0,0)
probe_1 = pwp.generate_sinc_probe(Nact, probe_amp, probe_radius, probe_phase=0, offset=probe_offset, bad_axis='x')
probe_2 = pwp.generate_sinc_probe(Nact, probe_amp, probe_radius, probe_phase=2*np.pi/3, offset=probe_offset, bad_axis='y')
probe_3 = pwp.generate_sinc_probe(Nact, probe_amp, probe_radius, probe_phase=4*np.pi/3, offset=probe_offset, bad_axis='x')
# probe_4 = pwp.generate_sinc_probe(Nact, amp, probe_radius, probe_phase=3*np.pi/4, offset=(6,12), bad_axis='y')

probes = np.array([probe_1.flatten(), probe_2.flatten(), probe_3.flatten()])

misc.myimshow(probe_1)
misc.myimshow(probe_2)
misc.myimshow(probe_3)
# misc.myimshow(probe_4)

# probe_modes_1 = iefcu.create_probe_poke_modes(Nact, 1*Nact//4, 2*Nact//3, 3*Nact//4-1, 2*Nact//3)
# probe_modes_2 = iefcu.create_probe_poke_modes(Nact, 1*Nact//2, 1*Nact//4, 3*Nact//4-1, Nact//2)
# probe_modes = np.vstack((probe_modes_1, probe_modes_2[0]))

# probes = probe_amp * probe_modes

# misc.myimshow((probes[0] + probes[1] + probes[2]).reshape((Nact,Nact)))
```

```python
probe_1_field = cp.exp(1j* cp.array(probes[2].reshape(Nact,Nact)) /hlci.wavelength_c.value * 2*cp.pi)
probe_1_field = cp.array(probes[2].reshape(Nact,Nact))
E_probe_1 = cp.fft.ifftshift(cp.fft.fft2(cp.fft.fftshift(probe_1_field)))
misc.myimshow2(cp.abs(E_probe_1), cp.angle(E_probe_1))
```

```python
hlci = cgi.CGI_POPPY(polaxis=0, use_opds=True, use_fieldstop=True, quiet=True, ngpus=1/2)
hlci.set_dm1(2*dm1_flatmap)
hlci.set_dm2(2*dm2_flatmap)

dm1_ref = hlci.DM1.surface.flatten().get()
dm2_ref = hlci.DM2.surface.flatten().get()

dm_commands_p = []
dm_commands_m = []
for i in range(len(probes)):
    dm_commands_p.append(np.vstack((probes[i] + dm1_ref, dm2_ref)))
    dm_commands_m.append(np.vstack((-probes[i] + dm1_ref, dm2_ref)))
    
hlci.reset_dms()
wfs_p = hlci.calc_psfs(dm_commands=dm_commands_p)
hlci.reset_dms()
wfs_m = hlci.calc_psfs(dm_commands=dm_commands_m)

Ip = []
Im = []
for i in range(len(probes)):
    Ip.append(wfs_p[i][-1].intensity)
    Im.append(wfs_m[i][-1].intensity)
    
    misc.myimshow2(Ip[i], Im[i], lognorm1=True, lognorm2=True)
    misc.myimshow(Ip[i]-Im[i])
    
E_probes = cp.zeros((probes.shape[0], 2*nmask))
I_diff = cp.zeros((probes.shape[0], nmask))
for i in range(len(probes)):
    E_probe = jac.dot(cp.array(probes[i]))
    E_probe = E_probe[:nmask] + 1j*E_probe[nmask:]
    
    E_probes[i, :nmask] = E_probe.real
    E_probes[i, nmask:] = E_probe.imag
    
    I_diff[i:(i+1), :] = (Ip[i] - Im[i]).flatten()[dh_mask]
```

```python
# use batch process
E_est = cp.zeros((nmask,), dtype=cp.complex128)
for i in range(nmask):
    delI = I_diff[:, i]
    M = 2*cp.array([E_probes[:,i], E_probes[:,i+nmask]]).T
    
    Minv = iefcu.TikhonovInverse(M, 1e-2)
    
    est = Minv.dot(delI)
    
    E_est[i] = est[0] + 1j*est[1]

E_est_2d = cp.zeros((npsf,npsf), dtype=cp.complex128)
cp.place(E_est_2d, mask=dh_mask, vals=E_est)
misc.myimshow2(cp.abs(E_est_2d)**2, cp.angle(E_est_2d), 
               'Estimated Intensity', 'Estimated Phase', lognorm1=True)
```
