```python
import numpy as np
import cupy as cp
import astropy.units as u
from astropy.io import fits
from matplotlib.patches import Rectangle, Circle
from pathlib import Path
from IPython.display import clear_output
from importlib import reload
import time

import logging, sys
poppy_log = logging.getLogger('poppy')
poppy_log.setLevel('INFO')
logging.basicConfig(stream=sys.stdout, level=logging.INFO)

poppy_log.disabled = True

import warnings
warnings.filterwarnings("ignore")

import hcipy as hci

from wfsc import iefc_2dm as iefc
from wfsc import utils

import misc_funs as misc

wavelength = 575e-9*u.m
dm_diam = 10.2*u.mm
f = 500*u.mm

dm1_dm2 = 200e-3*u.m
FN = (dm_diam**2/dm1_dm2/wavelength).decompose()
z_talbot = (dm_diam**2/2/wavelength).decompose()
display(dm1_dm2, z_talbot, FN)

wave_m = wavelength.to_value(u.m)
dm_diam_m = dm_diam.to_value(u.m)
dm1_dm2_m = dm1_dm2.to_value(u.m)
f_m = f.to_value(u.m)
```

```python
# define the grids
npix = 256
oversample = 2
spatial_resolution = f_m * wave_m / dm_diam_m

pupil_grid = hci.make_pupil_grid(npix*oversample, dm_diam_m * oversample)
focal_grid = hci.make_focal_grid(2, 16, spatial_resolution=spatial_resolution)
prop = hci.FraunhoferPropagator(pupil_grid, focal_grid, f_m)

aperture = hci.Field(np.exp(-(pupil_grid.as_('polar').r / (0.5 * dm_diam_m))**30), pupil_grid)
```

```python
# make the aberrations of the system (make sure to remove tip-tilt)
aberration_ptv = 0.02 * wave_m
tip_tilt = hci.make_zernike_basis(3, dm_diam_m, pupil_grid, starting_mode=2)
wfe = hci.SurfaceAberration(pupil_grid, aberration_ptv, dm_diam_m, remove_modes=tip_tilt, exponent=-3)

aberration_distance_m = 100e-3
wfe_at_distance = hci.SurfaceAberrationAtDistance(wfe, aberration_distance_m)

wfe = wfe_at_distance(hci.Wavefront(aperture, wave_m))

misc.imshow2(wfe.intensity.shaped, wfe.phase.shaped, 
             vmin2=-0.5, vmax2=0.5)
```

```python
Nact = 34
actuator_spacing = 1.05 / Nact * dm_diam_m

influence_functions = hci.make_xinetics_influence_functions(pupil_grid, Nact, actuator_spacing)

dm1 = hci.DeformableMirror(influence_functions)
dm2 = hci.DeformableMirror(influence_functions)

prop_between_dms = hci.FresnelPropagator(pupil_grid, dm1_dm2_m)

coronagraph = hci.PerfectCoronagraph(aperture, order=6)
```

```python
import poppy
class CORO():

    def __init__(self, 
                 wavelength=None, 
                ):
        
        self.wavelength_c = 575e-9*u.m
        self.wavelength = self.wavelength_c if wavelength is None else wavelength
        self.dm_diam = 10.2*u.mm
        self.f = 500*u.mm

        self.dm1_dm2 = 200e-3*u.m
        self.FN = (self.dm_diam**2/self.dm1_dm2/self.wavelength).decompose()
        self.z_talbot = (self.dm_diam**2/2/self.wavelength).decompose()

        self.wave_m = self.wavelength.to_value(u.m)
        dm_diam_m = dm_diam.to_value(u.m)
        dm1_dm2_m = dm1_dm2.to_value(u.m)
        f_m = f.to_value(u.m)
        
        # define the grids
        self.npix = 256
        self.oversample = 2
        self.spatial_resolution = f_m * self.wave_m / dm_diam_m

        pupil_grid = hci.make_pupil_grid(self.npix*self.oversample, dm_diam_m * self.oversample)
        focal_grid = hci.make_focal_grid(2, 16, spatial_resolution=self.spatial_resolution)
        self.prop = hci.FraunhoferPropagator(pupil_grid, focal_grid, f_m)

        aperture = hci.Field(np.exp(-(pupil_grid.as_('polar').r / (0.5 * dm_diam_m))**30), pupil_grid)
        
        # define the DMs
        self.Nact = 34
        actuator_spacing = 1.00 / Nact * dm_diam_m

#         influence_functions = hci.make_xinetics_influence_functions(pupil_grid, Nact, actuator_spacing)

        self.DM1 = hci.DeformableMirror(influence_functions)
        self.DM2 = hci.DeformableMirror(influence_functions)

        self.prop_between_dms = hci.FresnelPropagator(pupil_grid, dm1_dm2_m)
        
        self.dm_mask = np.ones((self.Nact,self.Nact), dtype=bool)
        xx = (np.linspace(0, self.Nact-1, self.Nact) - self.Nact/2 + 1/2) * actuator_spacing*2
        x,y = np.meshgrid(xx,xx)
        r = np.sqrt(x**2 + y**2)
        self.dm_mask[r>0.0105] = 0 # had to set the threshold to 10.5 instead of 10.2 to include edge actuators
        
        self.dm_zernikes = poppy.zernike.arbitrary_basis(self.dm_mask, nterms=15, outside=0)
        
        # make the aberrations of the system (make sure to remove tip-tilt)
        aberration_ptv = 0.02 * wave_m
        tip_tilt = hci.make_zernike_basis(3, dm_diam_m, pupil_grid, starting_mode=2)
        wfe = hci.SurfaceAberration(pupil_grid, aberration_ptv, dm_diam_m, remove_modes=tip_tilt, exponent=-3)

        aberration_distance_m = 100e-3
        self.wfe_at_distance = hci.SurfaceAberrationAtDistance(wfe, aberration_distance_m)
        
        # make the coronagraph model
        self.coronagraph = hci.PerfectCoronagraph(aperture, order=6)
        
        
    def set_dm1(command):
        self.DM1.actuators = command.ravel()
        
    def set_dm2(command):
        self.DM2.actuators = command.ravel()
        
    def add_dm1(command):
        self.DM1.actuators += command.ravel()
        
    def add_dm2(command):
        self.DM2.actuators += command.ravel()
        
    def get_dm1(command):
        return self.DM1.actuators.reshape(self.Nact, self.Nact)
        
    def get_dm2(command):
        return self.DM2.actuators.reshape(self.Nact, self.Nact)
        
    def calc_psf(self):
        wf = hci.Wavefront(aperture, self.wave_m)
        wf = self.wfe_at_distance(wf) # apply aberrations to wavefront
        wf = self.prop_between_dms.backward(self.DM2(prop_between_dms.forward(self.DM1(wf)))) # apply DMs
        wf = self.prop(self.coronagraph(wf))
        psf_wf = wf.real.shaped + 1j*wf.imag.shaped

        return np.array(psf_wf)

    def snap(self):
        wf = hci.Wavefront(aperture, self.wave_m)
        wf = self.wfe_at_distance(wf) # apply aberrations to wavefront
        wf = self.prop_between_dms.backward(self.DM2(prop_between_dms.forward(self.DM1(wf)))) # apply DMs
        im = self.prop(coronagraph(wf))

        return np.array(im.intensity.shaped)
```

```python
c = CORO()
im = c.snap()

misc.imshow1(im, lognorm=True)
```
