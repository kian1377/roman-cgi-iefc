```python
from scipy.linalg import hadamard

def get_hadamard_modes_2(sysi):
    Nact = sysi.Nact
    Nacts = int(sysi.dm_mask.sum())
    Ntot = 2*Nacts
    print(Nacts, Ntot)

    np2 = 2**int(np.ceil(np.log2(Ntot)))
    hmodes = hadamard(np2)
    print(hmodes.shape)

    inds = np.where(sysi.dm_mask.flatten().astype(int))

    hmodes_1 = hmodes[::2]
    hmodes_2 = hmodes[1::2]

    dm1_hads = []
    dm2_hads = []
    for i in range(hmodes_1.shape[0]):
        dm1_had = np.zeros((sysi.dm_mask.shape[0]**2))
        dm1_had[inds] = hmodes_1[i][:Nacts]

        dm2_had = np.zeros((sysi.dm_mask.shape[0]**2))
        dm2_had[inds] = hmodes_2[i][:Nacts]

        dm1_had = dm1_had.reshape(Nact, Nact)
        dm2_had = dm2_had.reshape(Nact, Nact)
        dm1_hads.append(dm1_had.flatten())
        dm2_hads.append(dm2_had.flatten())
        
    print(np.array(dm1_hads).shape)
    had_modes = np.concatenate((dm1_hads,dm2_hads), axis=1)
    print(had_modes.shape)
    return had_modes
had_modes = get_hadamard_modes_2(sysi)[:1024]

i = 10
dm1 = had_modes[i,:Nact**2].reshape(Nact, Nact)
dm2 = had_modes[i,Nact**2:].reshape(Nact, Nact)
print(dm1.flatten().dot(dm2.flatten()))
misc.myimshow3(dm1, dm2, dm1-dm2)
```

```python
def get_hadamard_modes_2(Nacts, dm_mask): 
    np2 = 2**int(np.ceil(np.log2(Nacts)))
    hmodes = hadamard(np2)
    print(hmodes.shape)
    
    Nact = int(dm_mask.sum())
    
    inds = np.where(dm_mask.flatten().astype(int))
    for hmode in hmodes:
        hmode_1 = hmode[:Nact]
        hmode_2 = hmode[Nacts//2:Nact]
        mode_1 = np.zeros((dm_mask.shape[0]**2))
        mode_1[inds] = hmode_1
        mode_2 = np.zeros((dm_mask.shape[0]**2))
        mode_2[inds] = hmode_2
        had_modes_1.append(mode_1)
        had_modes_2.append(mode_2)
    had_modes_1 = np.array(had_modes_1)
    had_modes_2 = np.array(had_modes_2)
    
    return had_modes_1, had_modes_2
```

```python
# def construct_control_matrix(response_matrix, 
#                              weight_map, 
#                              nc1, nc2, 
#                              nprobes=2, 
#                              rcond1=1e-2, 
#                              rcond2=1e-2, 
#                              WLS=True, 
#                              pca_modes=None):
#     weight_mask = weight_map>0
    
#     masked_matrix = response_matrix[:, :, weight_mask].reshape((response_matrix.shape[0], -1)).T
#     print(masked_matrix.shape)
    
#     nmodes = int(response_matrix.shape[0]/2)
#     if WLS:
#         print('Using Weighted Least Squares ')
#         if nprobes==2:
#             Wmatrix = np.diag(np.concatenate((weight_map[weight_mask], weight_map[weight_mask])))
#         elif nprobes==3:
#             Wmatrix = np.diag(np.concatenate((weight_map[weight_mask], weight_map[weight_mask], weight_map[weight_mask])))
#         control_matrix_1 = utils.WeightedLeastSquares(masked_matrix[:,:nc1], Wmatrix, rcond=rcond1)
#         control_matrix_2 = utils.WeightedLeastSquares(masked_matrix[:,nc1:], Wmatrix, rcond=rcond2)
#     else: 
#         print('Using Tikhonov Inverse')
#         control_matrix_1 = utils.TikhonovInverse(masked_matrix[:,:nc1], rcond=rcond1)
#         control_matrix_2 = utils.TikhonovInverse(masked_matrix[:,nc1:], rcond=rcond2)
        
#     control_matrix = np.concatenate((control_matrix_1, control_matrix_2), axis=0)
    
#     if pca_modes is not None:
#         # Return the control matrix minus the pca_mode coefficients
#         return control_matrix[0:-pca_modes.shape[0]]
#     else:
#         return control_matrix
    
def construct_control_matrix(response_matrix, 
                             weight_map, 
                             nprobes=2, 
                             rcond=1e-2,
                             WLS=True, 
                             pca_modes=None):
    weight_mask = weight_map>0
    
    masked_matrix = response_matrix[:, :, weight_mask].reshape((response_matrix.shape[0], -1)).T
    print(masked_matrix.shape)
    
    nmodes = int(response_matrix.shape[0])
    if WLS:
        print('Using Weighted Least Squares')
        if nprobes==2:
            Wmatrix = np.diag(np.concatenate((weight_map[weight_mask], weight_map[weight_mask])))
        elif nprobes==3:
            Wmatrix = np.diag(np.concatenate((weight_map[weight_mask], 
                                              weight_map[weight_mask], 
                                              weight_map[weight_mask])))
        control_matrix = utils.WeightedLeastSquares(masked_matrix[:,:nmodes], Wmatrix, rcond=rcond)
    else: 
        print('Using Tikhonov Inverse')
        control_matrix_1 = utils.TikhonovInverse(masked_matrix[:,:nc1], rcond=rcond1)
        control_matrix_2 = utils.TikhonovInverse(masked_matrix[:,nc1:], rcond=rcond2)
        
#     control_matrix = np.concatenate((control_matrix_1, control_matrix_2), axis=0)
    
    if pca_modes is not None:
        # Return the control matrix minus the pca_mode coefficients
        return control_matrix[0:-pca_modes.shape[0]]
    else:
        return control_matrix
```
